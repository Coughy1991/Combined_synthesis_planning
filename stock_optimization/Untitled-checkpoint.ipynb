```python
import rdkit
from rdkit import Chem
from rdkit.Chem import Descriptors
mol_lib = Chem.SDMolSupplier('/home/hanyug/Make-It/makeit/stock_optimization/real_targets/structures.sdf')
truncated_mol_lib = []
for mol in mol_lib:
    try:
        mw = Descriptors.ExactMolWt(mol)
        nhdonors = Chem.Lipinski.NumHDonors(mol)
        nhacceptors = Chem.Lipinski.NumHAcceptors(mol)
        logp = Chem.Crippen.MolLogP(mol)
        chiral_centers = Chem.FindMolChiralCenters(mol)
        ncarbons = 0
        for a in mol.GetAtoms():
            if a.GetSymbol() == 'C':
                ncarbons +=1
        if (sum([(mw>500), nhdonors>5, nhacceptors>10, logp>5])>=1) or len(chiral_centers)>3 or ncarbons<10:
            continue
        truncated_mol_lib.append(mol)
    except Exception as e: 
        print(e)
        pass

print(len(truncated_mol_lib))
```
