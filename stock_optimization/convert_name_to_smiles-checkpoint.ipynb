```python
###clean trees
import pickle
import numpy as np
with open('WHO_ban_rep/all_trees.pickle','r') as TREE_FILE:
    feasible_trees = pickle.load(TREE_FILE)
    
#### sort dataset and feasible trees
dataset_sorted = dataset.sort_values(['target','score'])
print(dataset_sorted.index)
feasible_trees = [feasible_trees[i] for i in list(dataset_sorted.index)]
dataset = dataset_sorted.reset_index(drop=True)
print(dataset.index)

def get_stepwise_fsc_from_tree(fsc=[],tree={}):
    if 'is_reaction' in tree.keys():
        fsc.append(tree['forward_score'])
    for child in tree['children']:
        get_stepwise_fsc_from_tree(fsc,child)
    return(fsc)
def nonzero_mul(x):
    prod = 1
    for element in x:
        if element>=0.05:
            prod = prod*element
    return(prod)
trees_to_keep = []
cum_score_list = []
cum_score_with_penalty = []
contains_zero = []
for i, tree in enumerate(feasible_trees):
    fsc = get_stepwise_fsc_from_tree(fsc=[],tree=tree['tree'])
    low_prob_count = sum([1 for sc in fsc if sc<0.05])
    if low_prob_count>1:
        continue
    elif low_prob_count==1:
        trees_to_keep.append(i)
        cum_score_list.append(nonzero_mul(fsc))
        cum_score_with_penalty.append(nonzero_mul(fsc)/2.71828)
        contains_zero.append(True)
    else: 
        trees_to_keep.append(i)
        cum_score_list.append(np.prod(fsc))
        cum_score_with_penalty.append(np.prod(fsc))
        contains_zero.append(False)

feasible_trees = [feasible_trees[i] for i in trees_to_keep]
# 
```

# Calculate SCScores for chemicals

```python
from makeit.prioritization.precursors.scscore import SCScorePrecursorPrioritizer
scscorer = SCScorePrecursorPrioritizer()
scscorer.load_model(model_tag='1024bool')

scscore_dict = {}
for key,value in chemicals_dict.iteritems():
    if 'context_' not in value:
        scscore_dict[key] = 1+scscorer.get_score_from_smiles(value, noprice=True)
    else:
        scscore_dict[key] = 1
```

```python
print(np.average([sc for sc in scscore_dict.values() if sc!=1]))
print(np.average([scscorer.get_score_from_smiles(target, noprice = True) for target in list(set(target_list))]))
```

```python
##########################PROBLEM 1, optimization with no context
####set up a problem with no context###########
from tqdm import tqdm
parameters_no_context = {}
route_counter = 0
target_list = []
for index,row in tqdm(dataset.iterrows()):
#     print(row['target'])
    if row['target'] not in target_list: ## if a new target, reset route counting
        parameters_no_context[row['target']]={}
#         print(parameters_no_context)
        #donot reset route_counter
#         route_counter = 0
        target_list.append(row['target'])
    parameters_no_context[row['target']][route_counter]={}
    for chemical in chemicals_list_index_no_context:
        parameters_no_context[row['target']][route_counter][chemical]=row[chemicals_dict_no_context[chemical]]
    route_counter+=1

    

from pulp import *

pathway_vars_no_context = LpVariable.dicts('target',target_list)

for target in target_list:
    routes_list = parameters_no_context[target]
    
    var_name = target + 'path'
    pathway_vars_no_context[target] = LpVariable.dicts(var_name,routes_list,cat=LpBinary)


chemical_vars_no_context = LpVariable.dicts('Chemical',chemicals_list_index_no_context, cat=LpBinary)
dummy_vars_no_context = LpVariable.dicts('Dummy',chemicals_list_index_no_context, cat=LpBinary)
prob_no_context = LpProblem('Pathway Optimization No Context', LpMinimize)
M=1000000
#objective
prob_no_context += lpSum([chemical_vars_no_context[i]*scscore_dict[i] for i in chemicals_list_index_no_context])

#constraints
# for any target, choose only one pathway
for target in target_list:
    prob_no_context += lpSum([pathway_vars_no_context[target][i] for i in pathway_vars_no_context[target]])==1

for chemical in tqdm(chemicals_list_index_no_context):
    prob_no_context += lpSum([lpSum([pathway_vars_no_context[target][i]*parameters_no_context[target][i][chemical] for i in pathway_vars_no_context[target]]) for target in target_list])<= M*dummy_vars_no_context[chemical]
    prob_no_context += chemical_vars_no_context[chemical] >= 1-M*(1-dummy_vars_no_context[chemical])

## if the pathway for any target contains a certain chemical, the final chemical pool includes that as well
print('problem is set up, start solving...')


prob_no_context.writeLP('stock_optimization_formulation_no_context.lp')
prob_no_context.solve(CPLEX())
```

```python
##########################PROBLEM 2, optimization with context
####set up a problem with no context###########

parameters = {}
route_counter = 0
target_list = []
for index,row in tqdm(dataset.iterrows()):
#     print(row['target'])
    if row['target'] not in target_list: ## if a new target, reset route counting
        parameters[row['target']]={}
#         print(parameters_no_context)
        #donot reset route_counter
#         route_counter = 0
        target_list.append(row['target'])
    parameters[row['target']][route_counter]={}
    for chemical in chemicals_list_index:
        parameters[row['target']][route_counter][chemical]=row[chemicals_dict[chemical]]
    route_counter+=1

    

from pulp import *

pathway_vars = LpVariable.dicts('target',target_list)

for target in target_list:
    routes_list = parameters[target]
    
    var_name = target + 'path'
    pathway_vars[target] = LpVariable.dicts(var_name,routes_list,cat=LpBinary)


chemical_vars = LpVariable.dicts('Chemical',chemicals_list_index, cat=LpBinary)
dummy_vars = LpVariable.dicts('Dummy',chemicals_list_index, cat=LpBinary)
prob = LpProblem('Pathway Optimization', LpMinimize)
M=1000000
#objective
prob += lpSum([chemical_vars[i]*scscore_dict[i] for i in chemicals_list_index])

#constraints
# for any target, choose only one pathway
for target in target_list:
    prob += lpSum([pathway_vars[target][i] for i in pathway_vars[target]])==1

for chemical in tqdm(chemicals_list_index):
    prob += lpSum([lpSum([pathway_vars[target][i]*parameters[target][i][chemical] for i in pathway_vars[target]]) for target in target_list])<= M*dummy_vars[chemical]
    prob += chemical_vars[chemical] >= 1-M*(1-dummy_vars[chemical])

## if the pathway for any target contains a certain chemical, the final chemical pool includes that as well
print('problem is set up, start solving...')


prob.writeLP('stock_optimization_formulation.lp')
prob.solve(CPLEX())
```

```python
##########################PROBLEM 3, optimization with context and evaluation
####set up a problem with no context###########
from tqdm import tqdm
parameters_eval = {}
parameters_score = {}
route_counter = 0
target_list = []
for index,row in dataset.iterrows():
#     print(row['target'])
##parameters for the scores of pathways
#     if row['score']>=0.5:
#         parameters_score[route_counter]= 0
#     else:
#         parameters_score[route_counter] = (0.5-row['score'])*10
#     parameters_score[route_counter] = (1-cum_score_list[index])*10  
    parameters_score[route_counter] = -np.log(cum_score_list[route_counter])+contains_zero[route_counter]
    if row['target'] not in target_list: ## if a new target, reset route counting
        parameters_eval[row['target']]={}
#         print(parameters_no_context)
        #donot reset route_counter
#         route_counter = 0
        target_list.append(row['target'])
    parameters_eval[row['target']][route_counter]={}
    parameters_eval[row['target']][route_counter]=dict(zip(chemicals_list_index,row[chemicals_list]))
    route_counter+=1
   
```

```python
 

from pulp import *

pathway_vars_eval = LpVariable.dicts('target',target_list)

for target in target_list:
    routes_list = parameters_eval[target]
    
    var_name = target + 'path'
    pathway_vars_eval[target] = LpVariable.dicts(var_name,routes_list,cat=LpBinary)


chemical_vars_eval = LpVariable.dicts('Chemical',chemicals_list_index, cat=LpBinary)
dummy_vars_eval = LpVariable.dicts('Dummy',chemicals_list_index, cat=LpBinary)
prob_eval = LpProblem('Pathway Optimization With Evaluation', LpMinimize)
M=1000000
#objective
prob_eval += lpSum([chemical_vars_eval[i]*scscore_dict[i] for i in chemicals_list_index]) \
+1*lpSum([lpSum([pathway_vars_eval[target][i]*parameters_score[i] for i in pathway_vars_eval[target]]) for target in target_list])

#constraints
# for any target, choose only one pathway
for target in target_list:
    prob_eval += lpSum([pathway_vars_eval[target][i] for i in pathway_vars_eval[target]])==1

for chemical in tqdm(chemicals_list_index):
    nonzero_paths = set(dataset[chemicals_dict[chemical]].nonzero()[0])
    prob_eval += lpSum([lpSum([(pathway_vars_eval[target][i]*parameters_eval[target][i][chemical]) for i in pathway_vars_eval[target] if i in nonzero_paths]) for target in target_list])<= M*dummy_vars_eval[chemical]
    prob_eval += chemical_vars_eval[chemical] >= 1-M*(1-dummy_vars_eval[chemical])

## if the pathway for any target contains a certain chemical, the final chemical pool includes that as well
print('problem is set up, start solving...')


prob_eval.writeLP('stock_optimization_formulation_eval.lp')
prob_eval.solve(CPLEX())
```
